```python
# inheritance 
'''
That's a very concise way to summarize the core of inheritance! You've got the parent-to-child access correct.
Can we access child attributes in the parent?

Short Answer: Generally, No, and you should avoid it.

Detailed Explanation:

Inheritance is defined as a one-way relationship .

    Parent knows nothing about the Child (The Rule):

        The Parent class is written before the Child class is ever defined.

        When you define the Parent class, the Child class (and its specific attributes, like Child.z) doesn't exist yet.

        Therefore, the Parent cannot contain any code that assumes or accesses a specific Child attribute.

    Why it's Bad Design:

        A single Parent might have ten different Child classes (e.g., Animal is the parent of Dog, Cat, and Bird).

        If the Animal class tried to access the Dog attribute leash_color, it would immediately 
        fail when you try to create a Cat or Bird object because they don't have that attribute.
        '''
```

```python
'''

Feature------------Instance Attribute-------------------------------------Class Attribute
Location-----------Lives on the specific object ("""""""self"""""").------Lives on the class blueprint itself.
Creation-----------Inside __init__ using self.name = value.---------------Direct assignment in the class body: name = value.
Sharing------------NO.Each object gets its own copy.----------------------YES. Shared by the class and all its objects.
Purpose------------"To hold data unique to that object--------------------To hold data common to all objects (e.g., Car.wheels = 4).
                    (e.g., car.color, cat.name).
C Equivalent-------A member field inside the struct itself.---------------A static variable shared by all structs of that type.

'''
```

```python
class animal:
    total_legs = 4 # Class attribute and YES. Shared by the class and all its objects.

    def __init__(self,name):
        self.name = name # Instance Attribute (Individual value)
    def change_total_legs(self, new_value):
        # CRUCIAL: Using 'self.total_legs' forces Python to create an 
        #    Instance Attribute, shadowing the Class Attribute for THIS object.
        self.total_legs = new_value

dog = animal("Fido")
cat = animal("whiskers")

print(f"Dog's legs: {dog.total_legs}")   # Output: 4 (Search 2)
print(f"Cat's legs: {cat.total_legs}")   # Output: 4 (Search 2)

print("now fido loses 1 leg")
dog.change_total_legs(3)
print(f"Dog's legs: {dog.total_legs}") 
print(f"Cat's legs: {cat.total_legs}")  
print(f"Class legs: {animal.total_legs}")
```

```python
'''
The Rule to Avoid Mistakes:

    To read a shared default: Use self.attribute.

    To write/change a shared default for everyone: Use the Class Name (animal.attribute = new_value).

    To change a value unique to the object: Use self.attribute = new_value (this creates the instance copy).

'''
```

```python
class Animal:

    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    
    # override eat() method
    def eat(self):
        # call the eat() method of the superclass using super()
        super().eat()
        print("I like to eat bones")
  

# create an object of the subclass
labrador = Dog()

labrador.eat()

#so main use of super is not replcaing it just additinally calls super class same func
#The primary use of super() is to additionally call the method of the superclass (parent) that has been overridden in the subclass (child).
```

```python
'''
Short Answer

Both prints run because you explicitly told Python to run both by including super().eat() inside the child's (Dog's) eat() method.

The purpose of super().eat() is not to replace the child's method, but to call the parent's version in addition to the child's specific logic.
Breakdown of the Execution

When you call labrador.eat(), this is the step-by-step process:

    The Child Method Starts: Python finds the eat() method in the Dog class and starts running it.

    Parent Method Call: The first line inside Dog.eat() is super().eat().

        This command pauses the Dog.eat() method and tells Python: "Go up one level in the inheritance hierarchy (to Animal) and execute its eat() method now."

    Parent Method Executes: The Animal.eat() method runs and executes its code:
    Python

print("I can eat")

(This produces the first printout.)

Control Returns to Child: Animal.eat() finishes, and control returns to the Dog.eat() method, right after the super().eat() line.

Child Method Continues: The Dog.eat() method executes its next line:
Python

print("I like to eat bones")

(This produces the second printout.)

'''
```

```python
'''

Both map() and filter() are built-in functions used to process lists and other iterables in a functional way, avoiding explicit loops.
1. map(function, iterable)

    Purpose: Transform ##every## element.

    Action: It takes a function and applies that function to #every single item# in the input list.

    Result: A new list containing the modified items.

        Example: Turn [1, 2, 3] into [10, 20, 30] (by multiplying by 10).

2. filter(function, iterable)

    Purpose: Select a subset of elements.

    Action: It takes a function that must return either True or False (a condition). It checks this condition against every item.

    Result: A new list containing only the items for which the function returned True.

        Example: Turn [1, 2, 3, 4] into [2, 4] (by selecting only even numbers).

'''
```

```python
'''
Purpose	Used in Generator Functions.	
Action	Pauses function execution and returns a value. State is saved.	
Output	A generator object (which is an iterator).	
Resumption	Resumes from where it left off on the next call.	
'''

# In contrast to return 
```

```python
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2
        
a= list(fibonacci_numbers(10))
b= list(square(fibonacci_numbers(10)))
print(a)
print(b)
print(sum(b))

```

```python
#The nonlocal keyword in Python is specifically used inside a nested function (or an inner function) 
#to indicate that a variable refers to a variable in the nearest enclosing scope that is not the global scope.
```

```python
def calculate():
    num = 1
    def inner_func():
        nonlocal num
        num += 2
        return num
    return inner_func

# call the outer function
odd = calculate()

# call the inner function
print(odd())
print(odd())
print(odd())

# call the outer function again
odd2 = calculate()
print(odd2())
```

```python
'''
Here is a step-by-step breakdown of what happens in your code:
1. odd = calculate()

    The calculate() function is executed for the first time.

    The variable num = 1 is created in calculate's local scope.

    calculate() returns the inner_func object.

    Crucially: Because inner_func references num (using nonlocal), Python recognizes that num must be kept alive. The returned inner_func object forms a closure over the scope of calculate(), permanently storing the value of num.

    The variable odd now holds this callable function (the closure).

    '''
```

```python
'''
Point,Description
1. Variable Persistence (The Core Effect),"When the outer function finishes executing,
its local variables that were referenced by the inner function are not destroyed. 
They are permanently preserved and bound to the inner function object."
2. Referencing vs. Modification (The Mechanism),"A closure is formed by the inner
function referencing (reading) the outer variable. If the inner function needs to modify 
that variable, you must explicitly use the nonlocal keyword to tell Python to update the preserved value, 
not create a new local one."
'''

```

```python
'''
If you have a single function (i.e., no nested functions) and that function simply returns a variable, then no, the variable is not kept "alive" in the memory after the function call is complete.
🛑 Why the Variable is Not Kept Alive:

    Local Scope Destruction: When a normal, non-nested function finishes executing, its local scope (where the variable was created) is destroyed.

    No Closure: A closure requires a nested function to capture the outer function's environment. Without nesting, there is no inner function to "carry" the variable and keep it alive.

    Return Value is a Copy/Reference: The return statement simply sends the value of the variable (or a reference to an object, like a list) back to the caller. It does not maintain the function's internal state.

📝 Example: No Nesting, No Persistence
'''
```

```python
def make_multiplier_of(n):
    def multiplier(x):
        return x * n
    return multiplier


# Multiplier of 3
times3 = make_multiplier_of(3) # HOLDS THE SPECIFIC CLOSURE THO with n=3 ..... WOW VAR HOLDS THE INSTANCE 

# Multiplier of 5
times5 = make_multiplier_of(5) 

# Output: 27
print(times3(9))

# Output: 15
print(times5(3))

# Output: 30
print(times5(times3(2)))
```

```python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "and", b)
        if b == 0:
            print("Whoops! cannot divide")
            return

        return func(a, b)
    return inner  # Instnace already returned think in terms of instance 

@smart_divide
def divide(a, b):
    print(a/b)

divide(2,5)

divide(2,0)
```

```python
def star(func):
    def inner(*args, **kwargs):
        print("start#" * 5)
        func(*args, **kwargs)
        print("S_END#" * 6)
    return inner


def percent(func):
    def inner(*args, **kwargs):
        print("inner_start#" * 5)
        func(*args, **kwargs)
        print("T_End%" * 6)
    return inner


@star
@percent
def printer(msg):
    print(msg)

printer("Hello")
```

```python
'''
n Python, we can pass a variable number of arguments to a function using special symbols. There are two special symbols:

    *args (Non Keyword Arguments)
    **kwargs (Keyword Arguments)

We use *args and **kwargs as an argument when we are unsure about the number of arguments to pass in the functions.
'''
```

```python
'''
In the function, we should use an asterisk * before the parameter name to pass variable length arguments.
The arguments are passed as a tuple and these passed arguments make tuple inside the function 
with same name as the parameter excluding asterisk *.

'''
```

```python
def intro(**data):
    print("\nData type of argument:",type(data))

    for key, value in data.items():
        print("{} is {}".format(key,value))

intro(Firstname="Sita", Lastname="Sharma", Age=22, Phone=1234567890)
intro(Firstname="John", Lastname="Wood", Email="johnwood@nomail.com", Country="Wakanda", Age=25, Phone=9876543210)
```

```python
def intro(**data):
    print("\nData type of argument:",type(data))

    for k in data:
        print(f"{k} is {data[k]}")

intro(Firstname="Sita", Lastname="Sharma", Age=22, Phone=1234567890)
```

```python
'''
If you have a list or tuple and want its elements to be 
passed sequentially by position, use the single * operator (for *args).

If you have a dictionary and want its keys to be used as 
argument names, use the double ** operator (for **kwargs).
'''
```

```python
def configure_user(**kwargs):
    if 'role' in kwargs:
        print(f"User Role: {kwargs['role']}")
    if 'status' in kwargs:
        print(f"Status: {kwargs['status']}")
    print(f"All keyword args (as dict): {kwargs}")

user_config = {'role': 'Editor', 'status': 'Active'}

# The double asterisk (**) unpacks the dictionary keys/values as KEYWORD arguments
configure_user(**user_config)
```
